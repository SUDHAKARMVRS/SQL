1. Retrieve the names of all customers who ordered products in the ‘Food’ category. Hint: Use JOIN between Customers, Orders, Order_Items, and Products.

```python
query = '''select c.customer_name,p.category from customers c 
 join orders o on c.customer_id = o.customer_id
 join order_items oi on o.order_id = oi.order_id
 join products p on oi.product_id = p.product_id
 where p.category = 'Food';'''
Food =pd.read_sql(query,engine)
Food
```

2. Find the total amount spent by each customer and display only those who spent more than 1000. Hint: Use aggregate SUM and GROUP BY with HAVING.

```python
query = '''select c.customer_name,sum(o.total_amount) as total_amt
from orders o join customers c on o.customer_id = c.customer_id
group by c.customer_name
having sum(total_amount)>1000'''
pro_orit = pd.read_sql(query,engine)
pro_orit
```

3. List customers who have never placed an order. Hint: Use LEFT JOIN and filter where Orders are NULL.

```python
query = '''select c.customer_name
 from customers c left join orders o on c.customer_id = o.customer_id
 left join order_items oi on oi.order_id = o.order_id
 where o.order_id is null'''
pro_nod = pd.read_sql(query,engine)
pro_nod
```

4. Show the top 3 selling products by total quantity sold. Hint: Join Order_Items and Products and use SUM with GROUP BY and ORDER BY.

```python
query = ''' select p.product_name,sum(oi.quantity) as total_sold_qty from order_items oi join products p on p.product_id = oi.product_id
group by p.product_name order by sum(oi.quantity) desc limit 3'''
pro_orit = pd.read_sql(query,engine)
pro_orit
```

5. Find customers who made payments using ‘UPI’ and placed orders after June 15, 2024. Hint: Join Payments and Orders and filter using DATE.

```python
query = '''select c.customer_name,o.order_date,p.payment_method
 from customers c join orders o on c.customer_id = o.customer_id
 join payments p on o.order_id = p.order_id
 where payment_method = 'UPI' and payment_date > '2024-06-15';'''
pro_nod = pd.read_sql(query,engine)
pro_nod
```

6. List all orders where the customer’s city name starts with ‘C’. Hint: Use JOIN and the LIKE operator.

9. List customers who ordered more than 5 different products. Hint: COUNT DISTINCT product_id grouped by customer_id.

```python
query = '''select c.customer_name,count(distinct(oi.product_id)) as total_product_count from customers c join orders o on c.customer_id = o.customer_id
join order_items oi on o.order_id = oi.order_id
join products p on oi.product_id = p.product_id
group by c.customer_id
having count(distinct(oi.product_id)) > 1
order by count(distinct(oi.product_id)) asc'''
total_product_ord = pd.read_sql(query,engine)
total_product_ord
```

22. List all customers who paid by ‘Card’ and bought ‘Dairy’ products. Hint: Use multiple JOINs and filter on payment_method and category.

```python
query = '''select c.customer_name,p.category 
from customers c 
join orders o on c.customer_id = o.customer_id
join order_items oi on o.order_id = oi.order_id
join products p on oi.product_id = p.product_id
join payments pm on o.order_id = pm.order_id
where pm.payment_method = 'Card' and p.category = 'Dairy';'''
cus_paym = pd.read_sql(query,engine)
cus_paym
```

23.Show product names with their total quantity sold and average quantity per order. Hint: Use SUM and AVG with GROUP BY product_id.

```python
query = '''select p.product_name,sum(oi.quantity) as total_qty,round(avg(oi.quantity),2) as avg_qty 
from products p
join order_items oi on p.product_id = oi.product_id
group by p.product_id
order by p.product_name asc'''
sum_avg = pd.read_sql(query,engine)
sum_avg
```
